# 5. Avg utilisation ratio vs churn :

```python
# Utilization Ratio Matters

# Higher average utilization ratio (spending more of their limit) might relate to churn (due to stress/debt).
```

```python
# Insight:
# 1.The presence of outliers in the attrited groups indicates that some churned customers have remarkably high utilisation ratios :

#2.Higher utlisation relate to churn:
# Higher utlisation means the customer is using most of their credit limit, possibly due to financial need or poor cash flow:
# These customers may struggle to pay back , leading to account closure or churn:



# 3.Risk Management by the Bank:
# Banks track utilization ratios to monitor risk. If a customer consistently uses 80â€“100% of their credit, the bank may flag them as risky and either reduce benefits or terminate the relationship.

# Customer Frustration with Limits:
# Customers hitting their credit limit often may get frustrated if their limits arenâ€™t raised, leading them to churn voluntarily.

# Fees and Interest Accumulation:
# High utilization often leads to higher interest charges or over-limit fees. This can lead to dissatisfaction and churn.

# Wider Spread:
# A wider spread among churned customers indicates inconsistency â€” some might churn due to low usage (inactive customers),
# while others churn due to very high usage (risk/stress).


# The bank should monitor customers with high utilization ratios and consider offering personalized credit counseling,
# limit adjustments, or rewards for healthy credit use to reduce churn risk.

```

```python
sns.boxplot(data=df, x='Attrition_Flag', y='Avg_Utilization_Ratio')

plt.title('Avg Utilization Ratio vs Churn')
plt.xlabel('Customer Churn (Attrition_Flag)')
plt.ylabel('Average Utilization Ratio')

plt.show()
```

```python
#  Insight: Transaction Count and Churn
#  Analysis of the Total_Trans_Ct (Total Transaction Count) shows that churned customers typically have lower
#  transaction activity compared to those who remain with the bank.

# Key Findings:

# Churned customers engage less frequently in transactions, indicating lower product engagement or inactive account behavior.

# Low usage may signal that the customer is not deriving enough value from the bankâ€™s services or has shifted to a competitor.

#  Recommendation:
# Customers showing a decline in transaction frequency over time should be targeted with re-engagement strategies, such as:

# Personalized offers

# Notifications of benefits

# Reminders about inactive services

#  Insight: Age and Churn Patterns
# From the Customer_Age feature, we observe that:

# Churn is higher in certain age brackets, particularly among middle-aged and older customers.

# This could reflect changing financial goals, reduced borrowing needs, or a preference for simpler banking experiences.

# Key Trends:

# Younger customers (e.g., 25â€“35) show higher engagement and lower churn.

# Customers over 55 may be more likely to churn, possibly due to retirement-related account changes or switching to conservative financial products.

# Recommendation:
# Develop age-tailored engagement strategies:

# For older customers: Offer simplified banking options, better interest products, or retirement services.

# For younger customers: Focus on digital services, rewards programs, and credit-building features.

# Would you like these insights compiled into a ready-to-use PDF or PowerPoint format summary?







```

```python
# 1 Find out the number of customer who are spending more than average_utliation ratio:
```

```python
attrited = df[df['Attrition_Flag'] == 'Attrited Customer']

Q1 = attrited['Avg_Utilization_Ratio'].quantile(0.25)
Q3 = attrited['Avg_Utilization_Ratio'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = attrited[(attrited['Avg_Utilization_Ratio'] < lower_bound) | 
                    (attrited['Avg_Utilization_Ratio'] > upper_bound)]

print("Number of customer who are spending more than avg utlisation ratio:", outliers.shape[0])
```

# 8. Customer Churn by Card Category:

```python
# Insight:
# person with the platinum card have the highest number of number of churned customer:
# person with the silver car have the lowest number of churn:
```

```python
card_category_churn = pd.crosstab(df['Card_Category'], df['Attrition_Flag'], normalize='index')

ax = card_category_churn.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set2')

for bar in ax.patches:
    width = bar.get_width()
    height = bar.get_height()
    x, y = bar.get_xy()
    
    if height > 0:
        ax.text(x + width / 2, y + height / 2,
                f'{height:.1%}',  # Convert to percentage
                ha='center', va='center', fontsize=10, color='black')

plt.title('Churn by Card Category')
plt.xlabel('Card Category')
plt.ylabel('Proportion of Customers')
plt.legend(title='Attrition Flag', loc='upper right')

plt.tight_layout()
plt.show()
```

```python
# Insight:
# person with the platinum card have the highest number of number of churned customer:(25%)
# person with the silver car have the lowest number of churn:(14.8%)
```
